### **Temperatura global: NASA GISS Surface Temperature Analysis (GISTEMP)**

El dataframe muestra el índice de temperatura global en la superficie de la Tierra y los océanos en unidades de 0.01 grados Celsius, basado en datos históricos de 1880 a 2024. Está organizado de la siguiente manera:

* **Year**: Año de la medición.
* **Columnas meses** : Cada columna desde enero hasta diciembre contiene la desviación de la temperatura media mensual global respecto al periodo base (1951-1980). Estos valores están expresados en centésimas de grados Celsius (por ejemplo, un valor de -20 indica -0.20 °C). 
* **Columnas de Promedios Anuales** : 
    - **J-D**: Promedio de temperatura de enero a diciembre de cada año.
    - **D-N**: Promedio de diciembre del año anterior hasta noviembre del año en curso.
    - **Invierno**: Promedio de diciembre del año anterior, enero y febrero del año en curso.
    - **Primavera**: Promedio de marzo, abril y mayo.
    - **Verano**: Promedio de junio, julio y agosto.
    - **Otoño**: Promedio de septiembre, octubre y noviembre.

Los datos se obtienen del sistema GHCN-v4 (Global Historical Climatology Network) y el índice de temperaturas oceánicas SST de ERSST v5. La serie de datos elimina valores atípicos y ajusta la homogeneidad para mantener consistencia.

*****: Este símbolo indica datos faltantes en casos donde no se tienen registros completos para un mes o año específico.

[Dataset Temperatura global](https://data.giss.nasa.gov/gistemp/tabledata_v4/GLB.Ts+dSST.txt)

```python
df_temperaturas = pd.read_csv("data\GLB.Ts+dSST.csv", header = 1)

df_temperaturas.rename(columns={
    'DJF': 'Invierno',
    'MAM': 'Primavera',
    'JJA': 'Verano',
    'SON': 'Otoño'
}, inplace=True)

df_temperaturas.replace('***', np.nan, inplace=True)
df_temperaturas.head()
```

### **Niveles de CO2: NOAA Global Monitoring Laboratory**

Este DataFrame es un conjunto de datos de concentraciones de dióxido de carbono (CO₂) en la atmósfera, proporcionado por el Laboratorio de Monitoreo Global (GML) de NOAA (Administración Nacional Oceánica y Atmosférica de Estados Unidos). Está organizado de la siguiente manera:

* **year**: Año en que se tomaron las mediciones.
* **month**: Indica el mes de la medición, con valores de 1 a 12, que corresponden a enero a diciembre respectivamente.
* **decimal**: Es una representación decimal del año, calculada para representar el mes de forma fraccional. Esto permite un seguimiento más continuo a lo largo del tiempo. Por ejemplo, 1979.042 representa enero de 1979, mientras que 1979.958 representa diciembre de 1979.
* **average**: Este valor indica la concentración promedio mensual de CO₂ en partes por millón (ppm) para ese mes específico. 
* **average_unc**: Muestra la incertidumbre asociada al valor promedio mensual de CO₂. Esta incertidumbre es el resultado de técnicas estadísticas, como el método Monte Carlo, que estiman la variabilidad al considerar diferentes conjuntos de registros de medición.
* **trend**: Representa una tendencia de concentración mensual de CO₂, ajustada para eliminar la variabilidad estacional (los cambios regulares que ocurren cada año debido a la vegetación y otras actividades naturales). Esta columna muestra una medida más estable y menos afectada por las fluctuaciones anuales de las estaciones.
* **trend_unc**: Muestra la incertidumbre asociada a la tendencia mensual del CO₂. Al igual que average_unc, indica el margen de error o la variabilidad en el cálculo de la tendencia mensual.

La columna trend es útil para analizar el cambio a largo plazo en la concentración de CO₂, eliminando las oscilaciones estacionales. La tendencia permite observar el aumento gradual de CO₂ en la atmósfera sin la interferencia de fluctuaciones estacionales.

Estos datos de CO₂ son cruciales para el monitoreo del cambio climático, ya que las concentraciones crecientes de CO₂ están vinculadas al calentamiento global. NOAA utiliza estos datos para analizar cómo varían las concentraciones de CO₂ en el tiempo y para proyectar tendencias futuras en la atmósfera terrestre.

### **Estructuramos de la misma forma el dataset de CO2 que el de Temperaturas**

```python
# Dataset para average por meses
df_CO2_average = df_CO2.pivot(index='year', columns='month', values='average').reset_index()
meses = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

df_CO2_average.rename(columns={1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}, inplace=True)
df_CO2_average = df_CO2_average.reset_index()
df_CO2_average.columns.name = None
df_CO2_average = df_CO2_average.iloc[:,1:]

# Calcular la media anual de CO₂
df_CO2_average['Annual_Avg'] = df_CO2_average.iloc[:, 1:].mean(axis=1).round(2)

df_CO2_average.head()
```

```python
# Dataset para average_unc por meses
df_CO2_average_unc = df_CO2.pivot(index='year', columns='month', values='average_unc').reset_index()
meses = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

df_CO2_average_unc.rename(columns={1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}, inplace=True)
df_CO2_average_unc = df_CO2_average_unc.reset_index()
df_CO2_average_unc.columns.name = None
df_CO2_average_unc = df_CO2_average_unc.iloc[:,1:]
df_CO2_average_unc.head()
```

```python
# Dataset para trend por meses
df_CO2_trend = df_CO2.pivot(index='year', columns='month', values='trend').reset_index()
meses = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

df_CO2_trend.rename(columns={1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}, inplace=True)
df_CO2_trend = df_CO2_trend.reset_index()
df_CO2_trend.columns.name = None
df_CO2_trend = df_CO2_trend.iloc[:,1:]
df_CO2_trend.head()
```

```python
# Dataset para trend_unc por meses
df_CO2_trend_unc = df_CO2.pivot(index='year', columns='month', values='trend_unc').reset_index()
meses = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

df_CO2_trend_unc.rename(columns={1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}, inplace=True)
df_CO2_trend_unc = df_CO2_trend_unc.reset_index()
df_CO2_trend_unc.columns.name = None
df_CO2_trend_unc = df_CO2_trend_unc.iloc[:,1:]
df_CO2_trend_unc.head()
```

El Eje Y indica cuánto se desvía la temperatura promedio global del periodo base. No son temperaturas absolutas, son valores que indican cuánto ha aumentado o disminuido la temperatura promedio de ese año en comparación del periodo 1951-1980 (periodo base).

La gráfica muestra una tendencia general ascendente, lo que indica que la temperatura promedio global ha ido aumentando entre 1983 y 2023. Si calculasemos el mínimo lo obtendríamos para el año 1983 y el máximo para el año 2023. Además podriamos calcular este aumento total, que sería 1.17-0.31 = 0.86, en cuarenta años.

Podemos observar que hay algunas fluctuaciones entre los años, con algunos periodos de enfriamiento temporal(descensos) seguidos de períodos de calentamiento. Esto es normal debido a la variabilidad climática natural, pero la tendencia a largo plazo muestra un aumento consistente en las temperaturas.

Esta tendencia ascendente es un indicador comúnmente observado en estudios de cambio climático, donde se evidencia el calentamiento global en función del tiempo.

### **El Niño y La Niña**

El Niño y La Niña son fases opuestas de un fenómeno climático conocido como el Oscilación del Sur de El Niño (ENSO, por sus siglas en inglés). Este fenómeno influye significativamente en el clima global y tiene efectos directos sobre las temperaturas anuales. 

Estos eventos son importantes porque introducen variabilidad natural en las temperaturas globales. Esto significa que, aunque hay una tendencia de calentamiento global a largo plazo, en los años con El Niño, las temperaturas pueden ser inusualmente altas, mientras que en los años con La Niña pueden ser más bajas. 

Estos fenómenos hay que tenerlos en cuenta para evitar interpretaciones erróneas de los datos de temperatura global y para distinguir entre la variabilidad natural y las tendencias de cambio climático inducidas por el ser humano.

* **El Niño**: Aumenta las temperaturas globales, causando años excepcionalmente cálidos y eventos climáticos extremos.

* **La Niña**: Disminuye las temperaturas globales, causando años ligeramente más fríos y fenómenos climáticos diferentes, como inviernos más fríos en algunas regiones.

#### **Registros Históricos de El Niño y La Niña**

Existen registros de cuándo han ocurrido estos eventos. Nos quedamos con los de intesidad fuerte.

* **Eventos de El Niño fuertes**: 1986-1987, 1997-1998, 2015-2016
* **Eventos de La Niña fuertes**: 1999-2000, 2010-2011

Vamos a marcar estos eventos en nuestra gráfica para ver si coinciden.

```python
el_nino_years = [1986,1987,1997, 1998, 2015, 2016]
la_nina_years = [1998, 2000, 2010, 2011]

plt.figure(figsize=(12, 6))
plt.plot(combined_df['Year'], combined_df['Annual_Mean'], color='green', label='Temperatura (°C)')
# Sombrear zonas de El Niño
for i in range(0, len(el_nino_years), 2):
    if i+1 < len(el_nino_years):
        plt.axvspan(el_nino_years[i], el_nino_years[i+1], color='lightblue', alpha=0.3, label='El Niño' if i == 0 else "")

# Sombrear zonas de La Niña
for i in range(0, len(la_nina_years), 2):
    if i+1 < len(la_nina_years):
        plt.axvspan(la_nina_years[i], la_nina_years[i+1], color='pink', alpha=0.3, label='La Niña' if i == 0 else "")


plt.title("Tendencia Anual de la Temperatura Promedio Global (1983-2023)  con Eventos El Niño y La Niña")
plt.xlabel("Año")
plt.ylabel("Temperatura (°C)")
plt.legend()
plt.grid(True)
plt.show()
```

Los períodos de El Niño (áreas rojas) coinciden con picos de temperatura global, lo cual demuestra el efecto de estos eventos en el calentamiento temporal de la atmósfera. Estos picos son evidentes en eventos como 1997-1998 y 2015-2016, que son algunos de los años más cálidos registrados.

Los períodos de La Niña (áreas azules) muestran un efecto moderador en la temperatura global, ayudando a reducir o estabilizar la temperatura en algunos años. Estos eventos no eliminan la tendencia general de calentamiento, pero pueden dar la impresión de "pausas" en algunos periodos.

Este gráfico muestra cómo los fenómenos de El Niño y La Niña introducen variabilidad en las temperaturas globales anuales. El Niño tiende a elevar las temperaturas, mientras que La Niña ayuda a reducirlas temporalmente. Sin embargo, la tendencia general es de aumento, lo que resalta el impacto del calentamiento global sobre el sistema climático terrestre.

### **Análisis por Décadas**

```python
def asignar_decada(year):
    if 1983 <= year <= 1992:
        return '1983-1992'
    elif 1993 <= year <= 2002:
        return '1993-2002'
    elif 2003 <= year <= 2012:
        return '2003-2012'
    elif 2013 <= year <= 2023:
        return '2013-2023'
    else:
        return None  

combined_df['Decada'] = combined_df['Year'].apply(asignar_decada)
```

```python
combined_df_groupby = combined_df.groupby("Decada")[["Annual_Mean","CO2_Annual_Avg"]].mean().reset_index()
combined_df_groupby
```

```python
plt.figure(figsize=(10, 6))
plt.plot(combined_df_groupby['Decada'], combined_df_groupby['Annual_Mean'], color='green', label='Temperaturas por Décadas')
plt.title("Tendencia por Décadas de la Temperatura Promedio")
plt.xlabel("Década")
plt.ylabel("Temperatura (°C)")
plt.grid(True)
plt.show()
```

En la Gráfica Anual de Temperatura (1983-2023) observabamos variaciones de un año a otro, con subidas y bajadas en ciertos periodos. Aunque la tendencia general es ascendente, podíamos observar algunos períodos de enfriamiento temporal seguidos de repuntes.

Al agrupar los datos por década, se suavizan las fluctuaciones anuales y se obtiene una línea más limpia y ascendente, que refleja la tendencia de calentamiento global a largo plazo. La tendencia por década muestra un aumento constante y casi lineal, lo que facilita visualizar la aceleración del calentamiento global sin distracciones de variabilidad anual.

```python
plt.figure(figsize=(10, 6))
plt.plot(combined_df_groupby['Decada'], combined_df_groupby['CO2_Annual_Avg'], color='blue', label='CO2 por Décadas')
plt.title("Tendencia por Décadas de la concentración promedio de CO2")
plt.xlabel("Década")
plt.ylabel("Concentración de CO₂ (ppm)")
plt.grid(True)
plt.show()
```

La Gráfica Anual de CO₂ (1983-2023) mostraba un aumento constante y suave en la concentración de CO₂ sin fluctuaciones notables.

La gráfica por década muestra una línea de crecimiento muy similar a la anual, pero con puntos más espaciados y un incremento claro en cada década. Al igual que la gráfica anual, la gráfica por década muestra que las concentraciones de CO₂ han aumentado de manera constante y se han acelerado en las últimas décadas.

```python
decadas = [(1983, 1992), (1993, 2002), (2003, 2012), (2013, 2023)]
correlaciones_decada = {}

for inicio, fin in decadas:
    df_decada = combined_df[(combined_df['Year'] >= inicio) & (combined_df['Year'] <= fin)]
    
    correlacion = df_decada['CO2_Annual_Avg'].corr(df_decada['Annual_Mean'])
    correlaciones_decada[f"{inicio}-{fin}"] = float(correlacion.round(2))

correlaciones_decada
```

```python
decadas = list(correlaciones_decada.keys())
correlaciones = list(correlaciones_decada.values())

plt.figure(figsize=(10, 6))
plt.bar(decadas, correlaciones, color=['skyblue', 'lightgreen', 'salmon', 'gold'])
plt.title("Correlación entre CO₂ Anual y Temperatura Promedio por Década")
plt.xlabel("Década")
plt.ylabel("Correlación")
plt.ylim(0, 1)
plt.grid(axis='y')

plt.show()
```

* **1983-1992** (0.52): Durante esta década, la correlación entre CO₂ y temperatura es moderada. Esto indica que, aunque hay una relación entre el aumento de CO₂ y el incremento de la temperatura, la relación no es muy fuerte.
* **1993-2002** (0.71): La correlación aumenta, lo que sugiere una relación más clara entre el aumento en los niveles de CO₂ y la temperatura. Esto podría reflejar que, en esta década, el impacto de las concentraciones de CO₂ en la temperatura se hizo más evidente.
* **2003-2012** (0.29): La correlación baja de manera significativa, lo cual sugiere que en esta década la relación entre el CO₂ y la temperatura fue más débil. Esto podría deberse a la influencia de otros factores climáticos.
* **2013-2023** (0.67): La correlación vuelve a aumentar, lo que indica nuevamente una fuerte relación entre CO₂ y temperatura. Esto es consistente con el calentamiento global observado en años recientes, donde las temperaturas tienden a subir en paralelo con los niveles de CO₂.

La relación entre CO₂ y temperatura no es constante y ha variado en cada década. Aunque hay una tendencia general de incremento en la correlación, la década 2003-2012 muestra una disminución, lo cual podría estar relacionado con variaciones climáticas no explicadas únicamente por el CO₂.

### **Análisis Comparativo entre average y trend**

Al comparar average con trend, es posible identificar años específicos en los que el CO₂ estuvo anormalmente alto o bajo. Estos años pueden ser investigados en mayor profundidad para entender las causas de las desviaciones.

```python
# Calcular las desviaciones entre el promedio y la tendencia
df_CO2['desviacion'] = (df_CO2['average'] - df_CO2['trend']).round(2)

df_CO2_anual = df_CO2.groupby('year').agg({
    'average': 'mean',
    'trend': 'mean',
    'desviacion': 'mean'
}).reset_index()


df_CO2_anual.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_CO2_anual['year'], df_CO2_anual['average'], label='CO₂ Promedio (average)', color='blue', linestyle='-', marker='o')
plt.plot(df_CO2_anual['year'], df_CO2_anual['trend'], label='CO₂ Tendencia (trend)', color='red', linestyle='--', marker='x')

plt.title("Comparación entre CO₂ Promedio (average) y Tendencia (trend) Agrupados por Año")
plt.xlabel("Año")
plt.ylabel("Concentración de CO₂ (ppm)")
plt.legend()
plt.grid(True)
plt.show()
```

Esta gráfica muestra la comparación entre la concentración promedio de CO₂ (average) y la tendencia suavizada de CO₂ (trend) a lo largo de los años.

* **Línea Azul (CO₂ Promedio - average)**: Representa la concentración promedio anual de CO₂. Esta línea muestra los datos reales de CO₂ año a año, incluyendo todas las fluctuaciones naturales y eventos climáticos que pueden afectar la concentración de CO₂ en la atmósfera.

* **Línea Roja Discontinua (CO₂ Tendencia - trend)**: Representa la tendencia suavizada de la concentración de CO₂. Esta línea elimina las variaciones de corto plazo y resalta el cambio a largo plazo, proporcionando una vista más clara del patrón general de incremento de CO₂.

Ambas líneas están muy cercanas entre sí, lo que sugiere que la tendencia general del CO₂ refleja fielmente la evolución del promedio anual.

Vamos a hacer un estudio más en profundidad sobre los años en los que el valor de CO₂ promedio (average) está significativamente por encima o por debajo de la tendencia (trend) encontrando los máximos y mínimos de la desviación entre ambas columnas.

Esto se debe a que la concentración de CO2 en la atmósfera presenta un patrón estacional que sigue un ciclo anual debido a la actividad fotosintética de las plantas y otros procesos naturales.

Durante el invierno (especialmente en el hemisferio norte), las plantas y los árboles están inactivos, por lo que la absorción de CO₂ por fotosíntesis es mínima. Además, la descomposición de la materia orgánica y otros procesos de respiración liberan CO₂. Esto provoca una acumulación de CO₂ en la atmósfera, que alcanza su punto máximo al final del invierno o principios de la primavera, justo antes de que las plantas comiencen a absorber CO₂ en gran cantidad.

En marzo y abril, las plantas comienzan su temporada de crecimiento, pero la absorción de CO₂ no es aún suficiente para contrarrestar la acumulación que se produjo durante el invierno. Por eso, en estos meses se observan los mayores valores de CO₂ atmosférico, hasta que la fotosíntesis alcanza su máxima capacidad en verano.
