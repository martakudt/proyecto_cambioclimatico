# **AUMENTO DE COMENTARIOS NEGACIONISTAS EN X**

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tuiter = {
    '2020': 220.000,
    '2021': 650.000,
    '2022': 850.000
}


df_tuiter = pd.DataFrame(list(tuiter.items()), columns=['Year', 'Tweets'])
```

```python
plt.figure(figsize=(8, 6))
sns.set_theme(style="whitegrid")

bar_plot = sns.barplot(
    data=df_tuiter,
    x='Year',
    y='Tweets',
    palette="Greens"
)

plt.title("Incremento de Tuits Negacionistas sobre el Cambio Climático", fontsize=14, weight='bold')
plt.xlabel("Año", fontsize=14)
plt.ylabel("Número", fontsize=14)

# Mostrar el gráfico
plt.show()
```

La gráfica muestra el incremento en el número de tuits negacionistas sobre el cambio climático entre los años 2020 y 2022. Los datos fueron obtenidos de un artículo de un periódico que analizó el crecimiento de este tipo de mensajes en redes sociales durante un periodo de tres años. En 2020, el número de tuits negacionistas era relativamente bajo, alrededor de 200 mil. Sin embargo, en 2021 se observa un aumento significativo, superando los 600 mil tuits. Este crecimiento continúa en 2022, donde el número de tuits alcanza los 800 mil.

La tendencia al alza es clara y preocupante, ya que indica un incremento continuo en la difusión de mensajes que niegan la existencia o el impacto del cambio climático. Este crecimiento en los mensajes negacionistas en redes sociales podría influir en la percepción pública sobre el cambio climático, generando desinformación y potencialmente afectando la voluntad de acción para abordar esta crisis global.
