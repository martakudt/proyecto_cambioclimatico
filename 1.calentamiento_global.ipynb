```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_invernadero = pd.read_csv("data/API_EN.GHG.ALL.MT.CE.AR5_DS2_en_csv_v2_10763.csv")
```

```python
df_invernadero = df_invernadero.drop(['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 68'], axis = 1)
```

```python
df_invernadero.columns
```

```python
df_invernadero = df_invernadero.loc[:,['Country Name','1983', '1984',
       '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993',
       '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002',
       '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011',
       '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020',
       '2021', '2022']]

df_invernadero.head()
```

```python
df_invernadero = df_invernadero.dropna(subset=['Country Name','1983', '1984',
       '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993',
       '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002',
       '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011',
       '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020',
       '2021', '2022'])


years_columns = ['1983', '1984',
       '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993',
       '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002',
       '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011',
       '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020',
       '2021', '2022']

df_invernadero
```

```python
df_invernadero['Total_Emissions'] = df_invernadero[years_columns].sum(axis=1)
df_invernadero['Mean Emissions'] = df_invernadero[years_columns].mean(axis=1).round(4)

df_sorted = df_invernadero.sort_values(by='Mean Emissions', ascending=False)
df_sorted = df_sorted.iloc[14:, :]

country_names = ['China', 'United States', 'European Union', 'India', 'Russian Federation', 'Japan', 'Brazil', 'Indonesia', 'Iran', 'Canada', 'Mexico', 'Saudi Arabia', 'Korea, Rep.', 'Australia', 'Turkiye', 'Pakistan' 'Thailand', 'Argentina', 'Kazakhstan']

df_sorted = df_sorted.loc[df_sorted.loc[:, "Country Name"].isin(country_names)]
df_sorted
```

```python
plt.figure(figsize=(12, 8))
sns.set_theme(style="whitegrid")


bar_plot = sns.barplot(
    data=df_sorted,
    x="Total_Emissions",
    y="Country Name",
    palette="Greens",
    edgecolor="black"
)


plt.title('PaÃ­ses con las mayores emisiones de gases de efecto invernadero (1983-2023)', fontsize=16, weight='bold')
plt.xlabel('Emisiones totales', fontsize=14)
plt.ylabel('PaÃ­s', fontsize=14)


plt.show()
```
