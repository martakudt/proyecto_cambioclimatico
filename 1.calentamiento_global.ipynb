# **EFECTO INVERNADERO**

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_invernadero = pd.read_csv("data/API_EN.GHG.ALL.MT.CE.AR5_DS2_en_csv_v2_10763.csv")
```

```python
df_invernadero = df_invernadero.drop(['Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 68'], axis = 1)
```

```python
df_invernadero.columns
```

```python
df_invernadero = df_invernadero.loc[:,['Country Name','1983', '1984',
       '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993',
       '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002',
       '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011',
       '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020',
       '2021', '2022']]

df_invernadero.head()
```

```python
df_invernadero = df_invernadero.dropna(subset=['Country Name','1983', '1984',
       '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993',
       '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002',
       '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011',
       '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020',
       '2021', '2022'])


years_columns = ['1983', '1984',
       '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993',
       '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002',
       '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011',
       '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020',
       '2021', '2022']

df_invernadero
```

```python
df_invernadero['Total_Emissions'] = df_invernadero[years_columns].sum(axis=1)
df_invernadero['Mean Emissions'] = df_invernadero[years_columns].mean(axis=1).round(4)

df_sorted = df_invernadero.sort_values(by='Mean Emissions', ascending=False)
df_sorted = df_sorted.iloc[14:, :]

country_names = ['China', 'United States', 'European Union', 'India', 'Russian Federation', 'Japan', 'Brazil', 'Indonesia', 'Iran', 'Canada', 'Mexico', 'Saudi Arabia', 'Korea, Rep.', 'Australia', 'Turkiye', 'Pakistan' 'Thailand', 'Argentina', 'Kazakhstan']

df_sorted = df_sorted.loc[df_sorted.loc[:, "Country Name"].isin(country_names)]
df_sorted
```

```python
plt.figure(figsize=(12, 8))
sns.set_theme(style="whitegrid")


bar_plot = sns.barplot(
    data=df_sorted,
    x="Total_Emissions",
    y="Country Name",
    palette="Greens",
    edgecolor="black"
)


plt.title('Países con las mayores emisiones de gases de efecto invernadero (1983-2023)', fontsize=16, weight='bold')
plt.xlabel('Emisiones totales', fontsize=14)
plt.ylabel('País', fontsize=14)


plt.show()
```

La gráfica muestra los países con las mayores emisiones de gases de efecto invernadero acumuladas entre 1983 y 2023, expresadas en "emisiones totales". Este gráfico de barras horizontales clasifica los principales emisores de gases de efecto invernadero y resalta la magnitud de sus emisiones en comparación con otros países.

* **Países Líderes en Emisiones**: China y Estados Unidos ocupan los dos primeros lugares, con una diferencia significativa respecto a otros países. La barra de China es la más larga, indicando que es el mayor emisor en este periodo. Le sigue Estados Unidos con un valor ligeramente inferior, pero aún considerablemente alto en comparación con otros países.
* **Contribuciones Significativas de Otras Regiones**: La Unión Europea, la Federación Rusa y la India también destacan en la gráfica, situándose en los primeros cinco puestos. Estos países y regiones tienen barras notablemente más largas que el resto, lo que indica que sus emisiones son considerablemente altas, aunque no alcanzan los niveles de China y Estados Unidos.
* **Otros Contribuyentes Relevantes**: Japón, Brasil e Indonesia se encuentran en una segunda categoría de emisores, con barras de tamaño medio en comparación con los primeros cinco países. Aun así, sus contribuciones son sustanciales en el contexto global. México, Canadá y Corea del Sur también aparecen en este grupo, aunque con barras más cortas.
* **Resto de los Emisores en el Top 15**: En la parte inferior de la lista, países como Australia, Arabia Saudita, Turquía, Argentina y Kazajistán también contribuyen a las emisiones globales, pero en menor proporción en comparación con los países en la parte superior de la gráfica. Sus barras son más cortas, indicando una contribución menor a las emisiones globales en este periodo de 40 años.
* **Tendencia General y Contexto**: La disposición de la gráfica y la magnitud de las barras reflejan que las emisiones de gases de efecto invernadero están altamente concentradas en unos pocos países y regiones. Este gráfico resalta el papel crucial que estos países desempeñan en la contribución al cambio climático, así como la importancia de que ellos lideren las políticas y acciones de reducción de emisiones para lograr un impacto global significativo.

En resumen, la gráfica muestra que, aunque muchos países contribuyen a las emisiones globales, la mayor parte de estas proviene de un grupo reducido de naciones.
